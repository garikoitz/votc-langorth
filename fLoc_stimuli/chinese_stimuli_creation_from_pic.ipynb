```python
import os
from PIL import Image, ImageDraw, ImageFont
import numpy as np


def get_image_files(directory):
    """List all image files in the votcloc/stimuli/scrambled."""
    supported_formats = ['.jpeg', '.jpg', '.png', '.bmp', '.gif']  # Add or remove formats as needed
    image_files = []
    for file in os.listdir(directory):
        if any(file.endswith(ext) for ext in supported_formats):
            image_files.append(os.path.join(directory, file))
    return image_files




def CN_word_transbg(image_path, output_dir):
    '''
    Get CN character from a grey background file
    input file size 308*308

    Output
    CN character in transparent background
    color can be set, default black
    '''
    
    image = Image.open(image_path).convert("RGBA")
    data = np.array(image)
    # Get the RGBA profile of this figure, threshold the region with grey background
    red, green, blue, alpha = data.T
    threshold = 127 
    grey_region = (red > threshold) & (green > threshold) & (blue > threshold)
    word_region=~grey_region

    data2=data

    data2[..., :-1][word_region.T] = (255, 255, 255)  # Set the color to black (can be adjusted)
    data2[..., -1][word_region.T] = 255  # Set the alpha channel to 0 (transparent)
    data2[..., :-1][grey_region.T] = (128, 128, 128)  # Set the color to black (can be adjusted)
    data2[..., -1][grey_region.T] = 0  # Set the alpha channel to 0 (transparent)
    result_image = Image.fromarray(data2)
    input_name= os.path.basename(image_path)
    cat= input_name.split('_')[0]
    output_name=input_name.replace(cat, 'TransBG-'+cat)
    result_image.save(os.path.join(output_dir,output_name))

    return result_image


input_CN_dir='/Users/tiger/toolboxes/votc-langorth/DATA/output/ZH_PNG'
CN_image_files= get_image_files(input_CN_dir)
output_dir='/Users/tiger/toolboxes/votc-langorth/DATA/CN_stim'

if not os.path.isdir(output_dir):
    os.makedirs(output_dir)
```

```python
def create_scrambled_figure(background_path, figure_path, output_path, tile_size=10):
    # Load the background and figure images
    background = Image.open(background_path)
    figure = Image.open(figure_path).convert("RGBA")  # Ensure figure is in RGBA mode

    # Get the dimensions of the figure and background
    fig_width, fig_height = figure.width, figure.height
    pic_width, pic_height = background.width, background.height

    # Calculate the number of tiles in each dimension based on tile_size
    num_tiles_x = fig_width // tile_size
    num_tiles_y = fig_height // tile_size

    # Create a list to store the tiles
    tiles = []

    # Extract tiles from the figure image
    for y in range(num_tiles_y):
        for x in range(num_tiles_x):
            left = x * tile_size
            upper = y * tile_size
            right = left + tile_size
            lower = upper + tile_size
            tile = figure.crop((left, upper, right, lower))
            tiles.append(tile)

    # Randomly shuffle the tiles
    random.shuffle(tiles)

    # Create a new image to reconstruct the scrambled image
    scrambled_image = Image.new("RGBA", (fig_width, fig_height), (255, 255, 255, 0))

    # Paste the shuffled tiles back into the scrambled image
    for y in range(num_tiles_y):
        for x in range(num_tiles_x):
            index = y * num_tiles_x + x
            left = x * tile_size
            upper = y * tile_size
            scrambled_image.paste(tiles[index], (left, upper))

    # Paste the scrambled figure onto the background image
    background.paste(scrambled_image, ((pic_width - fig_width) // 2, (pic_height - fig_height) // 2), scrambled_image)

    # Save the result
    background.save(output_path)
```

```python
import matplotlib.pyplot as plt
import os
join=os.path.join 
from PIL import Image
import numpy as np

def flip_image(image, split_point, image_name, output_dir):
    # Convert the image to RGBA and get its size
    width, height = image.size
    
    # Split the image into left and right halves at the selected split point
    left_half = image.crop((0, 0, split_point, height))
    right_half = image.crop((split_point, 0, width, height))
    
    # Create a new image where the left and right halves will be swapped
    flipped_image = Image.new("RGBA", (width, height))
    
    # Paste the right half on the left and the left half on the right
    flipped_image.paste(right_half, (0, 0))
    flipped_image.paste(left_half, (width - split_point, 0))
    
    flipped_image.save(join(output_dir,image_name.replace('.png','_flip.png')))
    return flipped_image

def onclick(event, image):
    # This function gets triggered when the user clicks on the image
    x_click = int(event.xdata)  # The x-coordinate of the click
    
    # Call the flip_image function to flip the image based on the click position
    flipped_image = flip_image(image, x_click)
    
    # Display the flipped image using matplotlib
    plt.imshow(np.array(flipped_image))
    plt.axis('off')  # Hide axes
    plt.show()

def interactive_image_split(image_path):
    # Load the image
    image = Image.open(image_path).convert("RGBA")
    
    # Display the image using matplotlib
    fig, ax = plt.subplots()
    ax.imshow(np.array(image))
    ax.set_title('Click to choose the split point')
    
    # Set up the event listener for mouse click
    cid = fig.canvas.mpl_connect('button_press_event', lambda event: onclick(event, image))
    
    plt.axis('off')  # Hide axes
    plt.show()

output_dir=join(homedir,'tlei/toolboxes/votc-langorth/DATA/CN_stim')
tp_img_dir=output_dir
image_path=os.path.join(tp_img_dir,'TransBG-CS_001_SHEN-01.png')
interactive_image_split(image_path)
```

```python
import cv2

class DrawLineWidget(object):
    def __init__(self):
        self.original_image = cv2.imread('1.jpg')
        self.clone = self.original_image.copy()

        cv2.namedWindow('image')
        cv2.setMouseCallback('image', self.extract_coordinates)

        # List to store start/end points
        self.image_coordinates = []

    def extract_coordinates(self, event, x, y, flags, parameters):
        # Record starting (x,y) coordinates on left mouse button click
        if event == cv2.EVENT_LBUTTONDOWN:
            self.image_coordinates = [(x,y)]

        # Record ending (x,y) coordintes on left mouse bottom release
        elif event == cv2.EVENT_LBUTTONUP:
            self.image_coordinates.append((x,y))
            print('Starting: {}, Ending: {}'.format(self.image_coordinates[0], self.image_coordinates[1]))

            # Draw line
            cv2.line(self.clone, self.image_coordinates[0], self.image_coordinates[1], (36,255,12), 2)
            cv2.imshow("image", self.clone) 

        # Clear drawing boxes on right mouse button click
        elif event == cv2.EVENT_RBUTTONDOWN:
            self.clone = self.original_image.copy()

    def show_image(self):
        return self.clone

if __name__ == '__main__':
    draw_line_widget = DrawLineWidget()
    while True:
        cv2.imshow('image', draw_line_widget.show_image())
        key = cv2.waitKey(1)

        # Close program with keyboard 'q'
        if key == ord('q'):
            cv2.destroyAllWindows()
            exit(1)
```

```python
import os
import shutil
import glob
src_dir='/home/tlei/tlei/toolboxes/votc-langorth/DATA/CN_stim'

def copy_and_rename_files(source_folder, destination_folder, pattern, rename_function):
    # Ensure the destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
    
    # Find all files that match the pattern
    files_to_copy = glob.glob(os.path.join(source_folder, pattern))
    
    # Copy and rename each file
    for file_path in files_to_copy:
        # Extract the file name
        file_name = os.path.basename(file_path)
        
        # Generate the new file name using the rename function
        new_file_name = rename_function(file_name)
        
        # Full path of the new destination
        new_file_path = os.path.join(destination_folder, new_file_name)
        
        # Copy the file to the new destination and rename it
        shutil.copy(file_path, new_file_path)
        print(f"Copied {file_name} to {new_file_name}")
# Example of a custom rename function
def rename_function(file_name):        # Split the file name by underscores and hyphen
    parts = file_name.split('-')
    
    # Extract the relevant parts (PW_060 in this case)
    new_name_part = parts[1]  # PW_060
    
    # Construct the new file name
    new_file_name = f"CN_{new_name_part}.png"
    return f"orig_{new_file_name}"
pattern = 'TransBG-CS_*.png'
copy_and_rename_files(src_dir,'/home/tlei/Desktop/CN_FF',pattern,rename_function)
```

## make CN Scamble

```python
def create_CN_SC(background_path, figure_path, output_path, tile_size=10):
    # Load the background and figure images
    background = Image.open(background_path)
    figure = Image.open(figure_path).convert("RGBA")  # Ensure figure is in RGBA mode

    # Get the dimensions of the figure and background
    fig_width, fig_height = figure.width, figure.height
    pic_width, pic_height = background.width, background.height

    # Calculate the number of tiles in each dimension based on tile_size
    num_tiles_x = fig_width // tile_size
    num_tiles_y = fig_height // tile_size

    # Create a list to store the tiles
    tiles = []

    # Extract tiles from the figure image
    for y in range(num_tiles_y):
        for x in range(num_tiles_x):
            left = x * tile_size
            upper = y * tile_size
            right = left + tile_size
            lower = upper + tile_size
            tile = figure.crop((left, upper, right, lower))
            tiles.append(tile)

    # Randomly shuffle the tiles
    random.shuffle(tiles)

    # Create a new image to reconstruct the scrambled image
    scrambled_image = Image.new("RGBA", (fig_width, fig_height), (255, 255, 255, 0))

    # Paste the shuffled tiles back into the scrambled image
    for y in range(num_tiles_y):
        for x in range(num_tiles_x):
            index = y * num_tiles_x + x
            left = x * tile_size
            upper = y * tile_size
            scrambled_image.paste(tiles[index], (left, upper))

    # Paste the scrambled figure onto the background image
    background.paste(scrambled_image, ((pic_width - fig_width) // 2, (pic_height - fig_height) // 2), scrambled_image)

    # Save the result
    background.save(output_path)

    src_dir='/home/tlei/Desktop'

homedir = os.getenv('HOME')
backgrounds_directory = join(homedir,"tlei/toolboxes/fLoc/stimuli/scrambled")
base_output_dir = join(homedir,"Desktop")

background_paths = get_image_files(backgrounds_directory)

src_type_dir=join(src_dir,f'CN_RW')
list_filenames=[join(src_type_dir,i) for i in os.listdir(src_type_dir)]
output_dir=join(src_dir,f'final_CN_SC')
os.makedirs(output_dir)
for pic in list_filenames:
    background_path = random.choice(background_paths)
    output_path=join(output_dir,pic.split('/')[-1]).replace('RW','SC')
    create_CN_SC(background_path, pic, output_path, tile_size=10)
```

```python
import re
def transform_filename(file_name):
    # Extract the relevant parts using a regular expression
    match = re.match(r'orig_(CN_[A-Z]{2})_(\d+)_.*\.png', file_name)
    
    if match:
        base_name = match.group(1)  # CN_RW
        number = int(match.group(2))  # Convert 001 to 1
        
        # Construct the new file name
        new_file_name = f"{base_name}-{number}.jpg"
        return new_file_name
    else:
        print(file_name)

def convert_jpg_to_png(file_path,output_dir):  
    img = Image.open(file_path)
    fname=file_path.split('/')[-1]
    # Get the new PNG file name using the transform_filename function
    new_file_name = transform_filename(fname)
    if new_file_name:
        # Create the full path for the JPEG file
        png_file_path = os.path.join(output_dir, new_file_name)
        
        # Convert and save the image as JPEG
        img.save(png_file_path, 'JPEG')
      
```
