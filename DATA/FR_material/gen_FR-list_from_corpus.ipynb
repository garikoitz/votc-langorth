```python
homedir = os.getenv('HOME')
word_dir = join(homedir,'toolboxes/votc-langorth/fLoc_stimuli/French/Lexique383')
fr_fname=join(word_dir,'Lexique383.tsv')
fr_df=pd.read_csv(fr_fname, sep="\t")
```

```python
# genre: in french seems we need to balance gender
# freqlemlivres: the frequency of the lemma in the books
# freqlivres: the frequency of the word in the books
# islemm: we choose the line that are 1
# nblettres: number of letters

# actually I will only get the lemma???
```

```python
# we want: 
# cgrams the catagory of grammer, we want NOM and VER, those will be the content words
# nblettres: 4,5,6
# islem: should be 1
# we want to use the freqlemlivres
# high frequency >200   
# low frequency < 30 https://link.springer.com/article/10.3758/BF03196834
```

```python
freq_limitation=(fr_df['freqlemlivres']>200) | ((fr_df['freqlemlivres']<20) & (fr_df['freqlemlivres']>5))
islem=fr_df['islem']==1
content_word=(fr_df.cgram=='VER')|(fr_df.cgram=='NOM')
length_456=fr_df['nblettres'].apply(lambda x: (x >= 4) and (x<=6))
```

```python
filtered_fr_df=fr_df[freq_limitation&islem&content_word&length_456].filter(items=['lemme','freqlemlivres']).drop_duplicates()
filtered_fr_df.to_csv(join(word_dir,'fr_filteted_5-20and_200.txt'), index=False)
```

```python
french_high=filtered_fr_df[filtered_fr_df['freqlemlivres']>200]
rand_high= random.sample(range(0,french_high.shape[0]), 100)
final_high_freq_list=french_high.iloc[rand_high]
final_high_freq_list.to_csv(join(word_dir,'FR_RWH.txt'), index=False)
```

```python
french_low=filtered_fr_df[filtered_fr_df['freqlemlivres']<20]
rand_low= random.sample(range(0,french_low.shape[0]), 130)
final_low_freq_list=french_low.iloc[rand_low]
final_low_freq_list.to_csv(join(word_dir,'FR_RWL.txt'), index=False)
```
